## GRAND QUESTION 1

__How does your name at your birth year compare to its use historically?__

_The name “nicholas” gradually increased in popularity until the mid 1970’s where the name became very popular and stayed very popular untill the early 2000’s where it saw a major decline in popularity._

```python
data.head()

felisha = data.query('name == "Felisha"')

felisha.sort_values(by='year').head()

```

```python
#| label: GQ1
#| code-summary: Read and format data

data.head()

data_birth = data.query('name ==  "Nicholas"').sort_values(by='year')

data_birth = data_birth[['name','year','Total']]

data_birth.head(20)
```

```python
#| label: GQ1 table
#| code-summary: table example
#| tbl-cap-location: top
alt.Chart(data_birth).mark_line().encode(x=alt.X('year',axis=alt.Axis(title='Year',format='Y')),
                                         y=alt.Y('Total',axis=alt.Axis(title='Total Times Name Used'))).properties(title='The name: "Nicholas" over time',
                                                                                                                   width=500,
                                                                                                                   height=500)
```

## GRAND QUESTION 2

__If you talked to someone named Brittany on the phone, what is your guess of his or her age? What ages would you not guess?__

_It’s very interesting to see the history of the name “Brittany” over time. If I were to encounter someone named Brittany, chances are they are between the age of 27-38 where the mid to early 30’s is especially likely. It’s very unlikely to meet some named Brittany who is 42 years old or older._


```python
#| label: GQ2
#| code-summary: Read and format data

data_brittany = data.query('name == "Brittany"').sort_values('year')

data_brittany['total_allTime'] = data_brittany['Total'].sum()

data_brittany['percent'] = round((data_brittany['Total'] / data_brittany['total_allTime']) * 100,4)

data_brittany['age'] = 2023 - data_brittany['year']

data_brittany = data_brittany[['name','year','age','percent']]

data_brittany.head(20)
```

```python
#| label: GQ2 chart
#| code-summary: plot example
#| fig-align: center
alt.Chart(data_brittany).mark_line().encode(x=alt.X('age',axis=alt.Axis(title="Brittany's Age")),
                                         y=alt.Y('percent',axis=alt.Axis(title='Percent Brittany was born during year'))).properties(title="Brittany's Age Predictions based on historical records",
                                                                                                                   width=500,
                                                                                                                   height=500)
```

## GRAND QUESTION 3

__Mary, Martha, Peter, and Paul are all Christian names. From 1920 - 2000, compare the name usage of each of the four names. What trends do you notice?__

_It looks like the name “Mary” was by far the most popular of the two names. All of the names showed a similar trend, during the late 1940’s to the 1960’s the names were all very popular and have been mostly a decline since._


```python
#| label: GQ3
#| code-summary: Read and format data
names = ['Mary','Martha','Peter','Paul']

data_names = data.query("1920 <= year <= 2000 and name in @names").sort_values(by=['name','year'])

data_names = data_names[['name','year','Total']]

data_names.head(20)
```

```python
#| label: GQ3 chart
#| code-summary: plot example
#| fig-align: center
alt.Chart(data_names).mark_line().encode(x=alt.X('year',axis=alt.Axis(title='Year',format='Y')),
                                         y=alt.Y('Total',axis=alt.Axis(title='Total Times Name Used')),
                                         color=alt.Color('name:N', legend=alt.Legend(title='Name')),
                                         tooltip=['year','Total','name']).properties(title='Christian names usage over time',width=500,height=500)
```
